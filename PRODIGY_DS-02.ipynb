```python
# Step 1: Load the dataset
import pandas as pd

# Load the Titanic dataset
titanic_df = pd.read_csv('C:/Users/movva/Downloads/test.csv')
print(titanic_df.head())

# Step 2: Data Cleaning
# Check for missing values
print(titanic_df.isnull().sum())

# Fill missing values
# Fill missing 'Age' values with the median age
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Fill missing 'Embarked' values with the most common port
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column as it has too many missing values
titanic_df.drop(columns=['Cabin'], inplace=True)

# Check for remaining missing values
print(titanic_df.isnull().sum())

# Step 3: Exploratory Data Analysis (EDA)
import seaborn as sns
import matplotlib.pyplot as plt

# Summary statistics
print(titanic_df.describe())

# Distribution of numerical features
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

sns.histplot(titanic_df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()

# Distribution of categorical features
sns.countplot(data=titanic_df, x='Pclass')
plt.title('Passenger Class Distribution')
plt.show()

sns.countplot(data=titanic_df, x='Sex')
plt.title('Gender Distribution')
plt.show()

sns.countplot(data=titanic_df, x='Embarked')
plt.title('Port of Embarkation Distribution')
plt.show()

# Survival rate by gender
sns.barplot(data=titanic_df, x='Sex', y='Survived')
plt.title('Survival Rate by Gender')
plt.show()

# Survival rate by class
sns.barplot(data=titanic_df, x='Pclass', y='Survived')
plt.title('Survival Rate by Class')
plt.show()

# Pairplot to visualize relationships between variables
sns.pairplot(titanic_df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']], hue='Survived')
plt.show()

# Correlation heatmap
corr_matrix = titanic_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
